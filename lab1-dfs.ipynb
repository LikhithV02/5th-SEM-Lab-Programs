```python
x, y, z = 12, 8, 5
mem = {}
ans = []
```

```python
def states(state):
    a = state[0]
    b = state[1]
    c = state[2]
    
    if a==6 and b==6:
        ans.append(state)
        return True
    
    if (a,b,c) in mem:
        return False
    mem[(a,b,c)] = 1
    
    #empty jug a
    if a>0:
        if a+b<=y:
            if states((0, a+b, c)):
                ans.append(state)
                return True
        else:
            if states((a-(y-b), y, c)):
                ans.append(state)
                return True
        if (a+c<=z):
            if states((0, b, a+c)):
                ans.append(state)
                return True
        else:
            if states((a-(z-c), b, z)):
                ans.append(state)
                return True
    #empty jug b
    if b>0:
        if a+b<=x:
            if states((a+b, 0, c)):
                ans.append(state)
                return True
        else:
            if states((x, b-(x-a), c)):
                ans.append(state)
                return True
        if (b+c<=z):
            if states((a, 0, b+c)):
                ans.append(state)
                return True
        else:
            if states((a, b-(z-c), z)):
                ans.append(state)
                return True
    #empty jug c
    if c>0:
        if a+c<=x:
            if states((a+c, b, 0)):
                ans.append(state)
                return True
        else:
            if states((x, b, c-(x-a))):
                ans.append(state)
                return True
        if (b+c<=y):
            if states((a, b+c, 0)):
                ans.append(state)
                return True
        else:
            if states((a, y, c-(y-b))):
                ans.append(state)
                return True
    return False
```

```python
ini = (12, 0, 0)
states(ini)
ans.reverse()
for i in ans:
    print(i)
```
