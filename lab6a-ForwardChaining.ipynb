```python
factskw = [['croaks', 'forg'], ['eats flies', 'forg'], ['frog', 'green'],
             ['chips', 'canary'], ['sing', 'canary'], ['canary', 'yellow']]
```

```python
def check(str, factOb):
    facts = []
    flag = True
    while flag:
        flag=False
        for txt in str:
            for A1 in factOb:
                if A1[0] == txt:
                    tmp = [txt, A1[1]]
                    if not tmp in facts:
                        facts += [tmp]
                        str += A1[1]
                        flag = True
    return facts
```

```python
result = check(['sing', 'canary'], factskw)
print(result)
```

```python
# Define the rule base
rules = {
    "rule1": {"premise": ["fact1", "fact2"], "conclusion": "fact3"},
    "rule2": {"premise": ["fact3"], "conclusion": "fact4"},
    "rule3": {"premise": ["fact2", "fact4"], "conclusion": "goal"}
}

# Define the initial fact base
fact_base = ["fact1", "fact2"]

# Define the goal
goal = "fact3"

# Function to apply a rule
def apply_rule(fact_base, rule):
    if all(fact in fact_base for fact in rule["premise"]):
        fact_base.append(rule["conclusion"])
        return True
    return False

# Function to perform forward chaining
def forward_chaining(fact_base, rules, goal):
    while goal not in fact_base:
        new_facts = []
        for rule in rules.values():
            if apply_rule(fact_base, rule):
                new_facts.append(rule["conclusion"])
                #print(new_facts)
        if not new_facts:
            return False
    return True

# Perform forward chaining
if forward_chaining(fact_base, rules, goal):
    print("Goal reached:", goal)
else:
    print("Goal not reached:", goal)
```
